# Infinite Row of Vortices 

For this exercise, we were asked to write code for an infinite row of vortices. Before attempting an infinite row of vortices, we programmed a code to calculate the velocity components and stream functions of a finite row of vortices through the use of functions and superposition. 

### Finite Row of Vortices 

First, we import the python libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

Create a grid mesh. 

```python
N = 50                          # number of grid points 
xstart,xend = -2.0,2.0          # x boundaries
ystart, yend = -1.0,1.0         # y boudaries 
x = np.linspace(xstart,xend,N)  # x direnction 1D array 
y = np.linspace(ystart,yend,N)  # y direction 1D array 
X,Y = np.meshgrid(x,y)          # mesh grid created 
```

Define functions to compute the velocity components of the vortices and their stream functions. 

```python
def getVelocityVortex(strength,xv,yv,X,Y):
    u = + strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v
    
def getStreamFunctionVortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi 
```

We need to initialize u, v, and psi by assigning zero arrays. Note, numpy arrays are used instead of lists because they are faster for future computations and require less memory allocation. Also, the strength of the vortex is the same for all so it is listed in this portion of the code as gamma. 

```python
uvortex=np.zeros((N,N),dtype=float)
vvortex=np.zeros((N,N),dtype=float) 
psivortex=np.zeros((N,N),dtype=float)

gamma = 5.0                   #strength of vortex (same for all) 
```

A fixed y-coordinate is assigned since it is a row only. The x-coordinate will vary but will be spaced evenly. For this exercise, the number of vortices selected was 9. 

```python
numbervortices = 9                               #number of vortices to be plotted 
xvortex=np.linspace(xstart,xend,numbervortices)  #location of vortices in the x-direction
yvortex = np.zeros(numbervortices)               #location of vortices in the y-direction  
```

A for loop is now applied to calculate u, v, and psi for every ith vortex. Here you can see how superposition is imposed as well. 

```python
for i in range(0,numbervortices):
    uvortex_init,vvortex_init = getVelocityVortex(gamma,xvortex[i],yvortex[i],X,Y)
    uvortex= uvortex + uvortex_init
    vvortex= vvortex + vvortex_init
    
    psivortex_init= getStreamFunctionVortex(gamma,xvortex[i],yvortex[i],X,Y)
    psivortex= psivortex + psivortex_init
```

Finally, we plot the results. 

```python
%matplotlib inline

size = 10
plt.figure(figsize=(size,(yend-ystart)/(xend-xstart)*size))
plt.xlim(xstart, xend)
plt.ylim(ystart, yend)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.title('Row of Finite Vortices',fontsize=20)
plt.streamplot(X,Y,uvortex,vvortex,\
              density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xvortex,yvortex,c='r',s=80,marker = 'o')
plt.show()
```

### Infinite Row of Vortices 

Now we attempt an infinite row of vortices. After some tedious math (refer to ipython lesson 05_Lesson05_InfiniteRowOfVortices), the equations for the u and v velocity component were obtained. We begin by defining the functions to compute the new velocity components for an infinite row .

```python
def getInfVelocityVortex(strength,a,X,Y):
    uinf = + strength/(2*a)*np.sinh(2*pi*Y/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    vinf = - strength/(2*a)*np.sin(2*pi*X/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    return uinf, vinf
```

The spacing between vortices is constant and will be defined as a0. Then the u and v are calculated using the function above. 

```python
a0 = (xend-xstart)/(numbervortices-1)
uinfvortex,vinfvortex=getInfVelocityVortex(gamma,a0,X,Y)
```

Now we can plot the infiite row of vortices. 

```python
size=10
plt.figure(figsize=(size, (yend-ystart)/(xend-xstart)*size))
plt.xlim(xstart, xend)
plt.ylim(ystart, yend)
plt.xlabel('x', fontsize = 16)
plt.ylabel('y', fontsize = 16)
plt.title('Infinite Row of Vortices', fontsize =22)
plt.streamplot(X,Y,uinfvortex,vinfvortex,\
                density = 2.0, linewidth = 1, arrowsize=1, arrowstyle='->')
plt.scatter(xvortex,yvortex,c='r',s=80,marker='o')
plt.show()
```
